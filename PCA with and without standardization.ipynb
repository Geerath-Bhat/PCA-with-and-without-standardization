```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('.\Physical_Feature_Data.csv')
```

```python
# Extract the features and target variable
X = df.drop('time', axis=1) # time is target variable here so we drop it from the features
y = df['time']
```

# PCA without standardization  

```python
# Compute the covariance matrix
cov_matrix = np.cov(X.T)
```

```python
# Compute the eigenvectors and eigenvalues of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

```python
# Sort the eigenvalues and eigenvectors in descending order
sort_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sort_indices]
eigenvectors = eigenvectors[:, sort_indices]
```

```python
# Choose the top 10 eigenvectors corresponding to the largest eigenvalues
top_eig_vecs = eigenvectors[:, :10]
```

```python
# Project the data onto the top 10 eigenvectors
PCA = np.dot(X, top_eig_vecs) # PCA without standarization
```

# PCA with standardization 

```python
# Standardize the data
data_std = (X - np.mean(X, axis=0)) / np.std(X, axis=0, ddof=1)
```

```python
data_std = data_std.drop('360_rot', axis=1)
```

```python
# Compute the covariance matrix
cov_matrix_std = np.cov(data_std.T)
```

```python
# Compute the eigenvectors and eigenvalues of the covariance matrix
eigvals, eigvect = np.linalg.eig(cov_matrix_std)
```

```python
# Sort the eigenvalues and eigenvectors in descending order
sort_indices = np.argsort(eigvals)[::-1]
eigvals = eigvals[sort_indices]
eigvect = eigvect[:, sort_indices]
```

```python
# Choose the top 10 eigenvectors corresponding to the largest eigenvalues
top_eig_vecs_std = eigvect[:, :10]
```

```python
# Project the data onto the top 10 eigenvectors
PCA_std = np.dot(data_std, top_eig_vecs_std) # PCA with standarization
```

# PCA with standardization plots

```python
# Plot the data points in the space of the 1st and 2nd principal components
plt.figure(figsize=(8,6))
plt.scatter(PCA_std[:,0], PCA_std[:,1], c='blue', alpha=0.5)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Data Points in the Space of PC1 and PC2')
```

```python
# Perform PCA on the 6th and 7th principal components
pcs_67 = PCA_std[:,5:7]

# Plot the data points in the space of the 6th and 7th principal components
plt.figure(figsize=(8,6))
plt.scatter(pcs_67[:,0], pcs_67[:,1], c='red', alpha=0.5)
plt.xlabel('PC6')
plt.ylabel('PC7')
plt.title('Data Points in the Space of PC6 and PC7')
```

```python
# Perform PCA on the 9th and 10th principal components
pcs_910 = PCA_std[:,8:10]

# Plot the data points in the space of the 9th and 10th principal components
plt.figure(figsize=(8,6))
plt.scatter(pcs_910[:,0], pcs_910[:,1], c='green', alpha=0.5)
plt.xlabel('PC9')
plt.ylabel('PC10')
plt.title('Data Points in the Space of PC9 and PC10')
```
